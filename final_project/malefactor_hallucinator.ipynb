## Model description
We will first train variational autoencoder on face images to compress them to low dimension. One important feature of VAE is that constructed latent space is dense. That means that we can traverse the latent space and reconstruct any point along our path into a valid face.

Using this continuous latent space we can use Bayesian optimization to maximize some similarity function between a person's face in victim/witness's memory and a face reconstructed from the current point of latent space. Bayesian optimization is an appropriate choice here since people start to forget details about the attacker after they were shown many similar photos. Because of this, we want to reconstruct the photo with the smallest possible number of trials.

## Search procedure

Now that we have a way to reconstruct images, we need to set up an optimization procedure to find a person that will be the most similar to the one we are thinking about. To do so, we need to set up some scoring utility. Imagine that you want to generate an image of Brad Pitt. You start with a small number of random samples, say 5, and rank them according to their similarity to your vision of Brad Pitt: 1 for the worst, 5 for the best. You then rate image by image using GPyOpt that works in a latent space of VAE. For the new image, you need to somehow assign a real number that will show how good this image is. The simple idea is to ask a user to compare a new image with previous images (along with their scores). A user then enters score to a current image.

The proposed scoring has a lot of drawbacks, and you may feel free to come up with new ones: e.g. showing user 9 different images and asking a user which image looks the "best".

````python
import bisect
import functools

class FacialComposit:
    def __init__(self, decoder, latent_size):
        self.latent_size = latent_size
        self.latent_placeholder = tf.placeholder(tf.float32, (1, latent_size))
        self.decode = decoder(self.latent_placeholder)
        self.samples = None
        self.images = None
        self.rating = None

    def _get_image(self, latent):
        img = sess.run(self.decode, 
                       feed_dict={self.latent_placeholder: latent[None, :]})[0]
        img = np.clip(img, 0, 1)
        return img

    @staticmethod
    def _show_images(images, titles):
        assert len(images) == len(titles)
        clear_output()
        plt.figure(figsize=(3*len(images), 3))
        n = len(titles)
        for i in range(n):
            plt.subplot(1, n, i+1)
            plt.imshow(images[i])
            plt.title(str(titles[i]))
            plt.axis('off')
        plt.show()

    @staticmethod
    def _draw_border(image, w=2):
        bordred_image = image.copy()
        bordred_image[:, :w] = [1, 0, 0]
        bordred_image[:, -w:] = [1, 0, 0]
        bordred_image[:w, :] = [1, 0, 0]
        bordred_image[-w:, :] = [1, 0, 0]
        return bordred_image
    
    @staticmethod
    def softmax(ranks, rank=None):
      return np.exp(ranks if rank is None else rank) / np.sum(np.exp(ranks))

    def query_initial(self, n_start=10, select_top=5):
        '''
        Creates initial points for Bayesian optimization
        Generate *n_start* random images and asks user to rank them.
        Gives maximum score to the best image and minimum to the worst.
        :param n_start: number of images to rank initialy.
        :param select_top: number of images to keep
        '''
        samples = [np.random.normal(size=[latent_size]) \
                   for _ in range(n_start)]
        images = [self._get_image(sample) for sample in samples]
        FacialComposit._show_images(images, ['Random Image'] * n_start)
        rank = input(f"Space seperated Rank. {n_start - 1} for best." \
                     " 0 for worst: ")
        rank = list(map(int, rank.split()))
        sample_rank = list(zip(samples, images, rank))
        sample_rank = sorted(sample_rank, key=lambda x: -x[2])
        samples, images, rank = list(zip(*sample_rank))
        rating = np.asarray(rank[:select_top])
        self.samples = np.asarray(samples[:select_top])
        self.images = np.asarray(images[:select_top])
        self.rating = np.exp(rating) / np.sum(np.exp(rating))
        np.testing.assert_equal(self.rating.shape, [select_top])
        np.testing.assert_equal(self.images.shape, [select_top, 64, 64, 3])
        np.testing.assert_equal(self.samples.shape, [select_top,
                                                     self.latent_size])

    def evaluate_softmax_renorm_rank(self, candidate):
        '''
        Queries candidate vs known image set.
        Adds candidate into images pool.
        Uses user input and softmax renormalization to estimate scores
        of all sampled images.
        :param candidate: latent vector of size 1xlatent_size
        '''
        
        initial_size = len(self.images)
        candidate_image = self._get_image(np.squeeze(candidate))
        FacialComposit._show_images([candidate_image] + list(self.images)[:3],
                                    ["Candidate Image"] + \
                                    ['Sample Images (Best First)'] * 3)
        print(candidate)
        rank = float(input("Enter a rank 0 if it has no match," \
                           " 1 if better than before or anything in between: "))
        assert rank >=0 and rank <= 1
        rank = rank * (np.max(self.rating) + 1e-20)
        renormalized_rank = FacialComposit.softmax(
            np.append(self.rating, rank), rank)
        self.rating = FacialComposit.softmax(np.append(self.rating, rank))[:-1]
        insert_idx = bisect.bisect_left(list(-self.rating), -renormalized_rank)
        rating_concat_list = [self.rating[:insert_idx],
                              renormalized_rank[np.newaxis],
                              self.rating[insert_idx:]]
        samples_concat_list = [self.samples[:insert_idx],
                               candidate, self.samples[insert_idx:]]
        images_concat_list = [self.images[:insert_idx],
                              candidate_image[np.newaxis, :],
                              self.images[insert_idx:]]
        if insert_idx >= len(self.rating):
          rating_concat_list.pop()
          samples_concat_list.pop()
          images_concat_list.pop()
        self.rating = np.concatenate(rating_concat_list)
        self.samples = np.concatenate(samples_concat_list)
        self.images = np.concatenate(images_concat_list)
        
        candidate_rating = renormalized_rank * 3
        assert len(self.images) == initial_size + 1
        assert len(self.rating) == initial_size + 1
        assert len(self.samples) == initial_size + 1
        return np.abs(candidate_rating) + 1e-20
    def check(self):
      return int(input('Enter 1 for better, 0 for worse: '))

    def evaluate_binary_estimate(self, candidate):
      '''
      Queries candidate vs known image set.
      Adds candidate into images pool.
      Uses binary search to estimate score of the worse sampled
      images.
      :param candidate: latent vector of size 1xlatent_size
      '''
      initial_size = len(self.images)
      candidate_image = self._get_image(np.squeeze(candidate))
      FacialComposit._show_images([candidate_image, self.images[0]],
                                  ['Candidate Image', "Current Image"])
      rank = self.check()
      images = list(self.images)
      samples = list(self.samples)
      rank_index = 0
      if rank:
        images.insert(0, candidate_image)
        rank_index = 0
        samples.insert(0, candidate)
      else:
        start = 0
        end = len(self.images) - 1
        print('Estimating Rank')
        while end > start:
          mid = (start + end) // 2
          FacialComposit._show_images([candidate_image, list(self.images)[mid]],
                                      ['Candidate Image', 'Current Image'])
          current_rank = self.check()
          if current_rank:
            mid -= 1
            end = mid
          else:
            mid += 1
            start = mid
        rank = np.random.uniform(0, self.rating[-1])
        if mid < initial_size - 1:
          rank = np.random.uniform(self.rating[mid + 1], self.rating[mid])
        rank_index = mid + 1
        images.insert(mid, candidate_image)
        samples.insert(mid, candidate)
      renormed_rank = self.softmax(np.append(self.rating, rank), rank)
      renormed_rating = list(self.softmax(np.append(self.rating, rank))[:-1])
      renormed_rating.insert(rank_index, renormed_rank)
      self.images = np.asarray(images)
      self.samples = np.asarray(samples)
      self.rating = np.asarray(renormed_rating)
      assert len(self.images) == initial_size + 1
      assert len(self.rating) == initial_size + 1
      assert len(self.samples) == initial_size + 1

      return renormed_rank
  
    def optimize(self, n_iter=10, w=4, acquisition_type='MPI',
                 acquisition_par=0.6, eval_fn=evaluate_softmax_renorm_rank):
        if self.samples is None:
            self.query_initial()

        bounds = [{'name': 'z_{0:03d}'.format(i),
                   'type': 'continuous',
                   'domain': (-w, w)}
                  for i in range(self.latent_size)]
        optimizer = GPyOpt.methods.BayesianOptimization(
            f=functools.partial(eval_fn, self), domain=bounds,
            acquisition_type = acquisition_type,
            acquisition_par = acquisition_par,
            exact_eval=False, # Since we are not sure
            model_type='GP',
            X=self.samples,
            Y=self.rating[:, None],
            maximize=True)
        optimizer.run_optimization(max_iter=n_iter)

    def get_best(self):
        index_best = 0
        return self.images[index_best]

    def draw_best(self, title=''):
        index_best = 0
        image = self.images[index_best]
        plt.imshow(image)
        plt.title(title)
        plt.axis('off')
        plt.show()
````

#### GRADED 2 (3 points):
Describe your approach below: How do you assign a score to a new image? How do you select reference images to help user assign a new score? What are the limitations of your approach?

The idea is to score the images based on the user's choice. For the inital query, The user is then asked to score the images with an integer score between [0, 5]. This a raw score, is then scaled down using softmax to keep everything within 0 and 1. The images with their corresponding latent vectors are then reverse sorted with respect to the softmax scores and stored in the list.
For the evalution scoring I was trying out two type of scoring systems.
- **In the First scoring technique:**  The user is shown the candidate image along with top two images to aid the comparison. Then the user is asked to score between [0, 1] to rate the next image. This raw score is then scaled and renormalized with respect to the previous ratings. This scaled and renormed score is then used to find the index where to insert the items in the previously sorted array, maintaining the order. (Done using bisection search). The scaled and renormed and score is then yield for the optimizer to work on.
- **For the second scoring technique**: The user is shown the candidate image and the best (highest score) entry of the previous entries. If the current candidate is better than the previous score, the item is added to as the current best and a new score is calculated by renormalizing the rating array after appending `1` to the `0th Index`. (Since 1 specifies better than previous and 0 specifies worse than before). If the current candidate is worse, a semi-automatic bisection search is performed on the previous rating array to to estimate the score. (Which is a random number from uniform distribution between the scores of the better image and the worse image with respect to the current candidate image)
